```python
print(soup.prettify())
```

```python
soup.find_all('h1')
```

```python
for tag_text in soup.find_all('h2'):
  print(tag_text.text.strip()) # using Strip to remove space and extrac characters
```

```python
for rating in soup.find_all('p', class_='rating'):
  print(rating.text.strip())
```

```python
for review in soup.find_all('a', class_='review-count'):
  print(review.text.strip())
```

```python
# whole company div

company = soup.find_all('div', class_='company-content-wrapper')
```

```python
name = []
rating = []
review = []
company_type=[]
hq = []
employess =[]
years_old = []

for i in company:
  name.append(i.find('h2').text.strip())
  rating.append(i.find('p', class_='rating').text.strip())
  review.append(i.find('a', class_='review-count').text.strip())
  company_type.append(i.find_all('p', class_='infoEntity')[0].text.strip())
  hq.append(i.find_all('p', class_='infoEntity')[1].text.strip())
  years_old.append(i.find_all('p', class_='infoEntity')[2].text.strip())
  try:
    employess.append(i.find_all('p', class_='infoEntity')[3].text.strip())
  except:
    employess.append(np.nan)

d = {'name': name, 'ratings':rating, 'reviews':review, 'company_type':company_type, 'hq_in':hq, 'total_employees':employess, 'years_old':years_old}

pd.DataFrame(d)
```

```python
# to loop through all pages
final = pd.DataFrame()

for page in range(1, 20):

  url = 'https://www.ambitionbox.com/list-of-companies?page={}'.format(page)

  headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
  webpage = requests.get(url, headers=headers).text

  soup = BeautifulSoup(webpage, 'lxml')

  company = soup.find_all('div', class_='company-content-wrapper')

  name = []
  rating = []
  review = []
  company_type=[]
  hq = []
  employess =[]
  years_old = []

  for i in company:
    name.append(i.find('h2').text.strip())
    rating.append(i.find('p', class_='rating').text.strip())
    review.append(i.find('a', class_='review-count').text.strip())
    try:
      company_type.append(i.find_all('p', class_='infoEntity')[0].text.strip())
    except:
      company_type.append(np.nan)
    try:
      hq.append(i.find_all('p', class_='infoEntity')[1].text.strip())
    except:
      hq.append(np.nan)
    try:
      employess.append(i.find_all('p', class_='infoEntity')[3].text.strip())
    except:
      employess.append(np.nan)
    try:
      years_old.append(i.find_all('p', class_='infoEntity')[2].text.strip())
    except:
      years_old.append(np.nan)

  d = {'name': name, 'ratings':rating, 'reviews':review, 'company_type':company_type, 'hq_in':hq, 'total_employees':employess, 'years_old':years_old}

  df = pd.DataFrame(d)

  final = final.append(df, ignore_index=True)
```
